# Logistic Regression

```python
dfResults =  pd.DataFrame()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


features = dfDroped.drop(columns=['RadiantWin'])
labels = df['RadiantWin']
test_size = 0.4
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = test_size)


model = LogisticRegression(solver='lbfgs', max_iter=100000)
model.fit(X_train, y_train)


# predict probabilities
pred_prob = model.predict_proba(X_train)


dfResults.loc[0,'Test_accuracy'] = model.score(X_test, y_test)
dfResults.loc[0,'Training_accuracy'] = model.score(X_train, y_train)
dfResults.loc[0,'test_size'] = test_size
dfResults
```

```python
for i in range(9):
    test_size = (i+1)/10   
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = test_size)
    model = LogisticRegression()
    model.fit(X_train, y_train)

    dfResults.loc[i,'Test_accuracy'] = model.score(X_test, y_test)
    dfResults.loc[i,'Training_accuracy'] = model.score(X_train, y_train)
    dfResults.loc[i,'test_size'] = test_size
dfResults
```
